# Desafio:

Você faz parte da equipe de Analytics de uma grande marca de vestuário com mais de 25 lojas espalhadas em Shoppings de todo o Brasil.

Toda semana você precisa enviar para a diretoria um ranking atualizado com as 25 lojas contendo 3 informações:
- Faturamento de cada Loja
- Quantidade de produtos vendidos de cada Loja
- Ticket Médio dos produtos de cada Loja

Além disso, cada loja tem 1 gerente que precisa receber o resumo das informações da loja dele. Por isso, cada gerente deve receber no e-mail:
- Faturamento da sua loja
- Quantidade de produtos vendidos da sua loja
- Ticket Médio dos produto da sua Loja

Esse relatório é sempre enviado como um resumo de todos os dados disponíveis no ano.

# Solução do Desafio:

Para resolver o desafio vamos seguir a seguinte lógica:

- Passo 1 - Importar a base de dados
- Passo 2 - Visualizar a base de dados para ver se precisamos fazer algum tratamento
- Passo 3 - Calcular os indicadores de todas as lojas:
  - Faturamento por loja
  - Quantidade de produtos vendidos por loja
  - Ticket Médio dos produto por loja
- Passo 4 - Calcular os indicadores de cada loja
- Passo 5 - Enviar e-mail para cada loja
- Passo 6 - Enviar e-mail para a diretoria

### Passo 1 - Importando a Base de Dados

```
import pandas as pd
df_sales = pd.read_excel(r'/content/drive/MyDrive/Colab Notebooks/Intensivão de Python/Aula 1/Sales.xlsx')
```

### Passo 2 - Visualizando os Dados

```
display(df_sales)
```

### Passo 3.1 - Calculando o Faturamento por Loja

```
billing = df_sales[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
billing = billing.sort_values(by='Valor Final', ascending=False)
display(billing)
```

### Passo 3.2 - Calculando a Quantidade Vendida por Loja

```
quantity = df_sales[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()
quantity = quantity.sort_values(by='ID Loja', ascending=False)
display(quantity)
```

### Passo 3.3 - Calculando o Ticket Médio dos Produtos por Loja

```
ticket_average = (billing['Valor Final'] / quantity['Quantidade']).to_frame()
ticket_average = ticket_average.rename(columns={0: 'Ticket Medio'})
ticket_average = ticket_average.sort_values(by='Ticket Medio', ascending=False)
display(ticket_average)
```

### Criando a Função de Enviar e-mail

```
import smtplib
import email.message

def send_email(shop_summary, shop):
  server = smtplib.SMTP('smtp.gmail.com:587')  
  email_content = f'''
  <p>Prezados,</p>
  {shop_summary.to_html()}
  <p>At.te,</p>
  <p>Bryan Lima</p>
  <p>Software Developer</p>
  '''
  
  msg = email.message.Message()
  msg['Subject'] = f'Relatório - {shop}'
    
  msg['From'] = SOURCE_EMAIL
  msg['To'] = DESTINATION_EMAIL

  msg.add_header('Content-Type', 'text/html')
  msg.set_payload(email_content)
  
  s = smtplib.SMTP('smtp.gmail.com: 587')
  s.starttls()
  
  s.login(msg['From'], PASSWORD_EMAIL)
  s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
```

### Passos 4 e 5 - Calculando Indicadores por Loja + Enviando E-mail para Cada Loja

```
shops = df_sales['ID Loja'].unique()

for shop in shops:
  table_shop = df_sales.loc[df_sales['ID Loja'] == shop, ['ID Loja', 'Quantidade', 'Valor Final']]
  shop_summary = table_shop.groupby('ID Loja').sum()
  shop_summary['Ticket Medio'] = shop_summary['Valor Final'] / shop_summary['Quantidade']
  send_email(shop_summary, shop)
```

### Passo 6 - Enviando E-mail para Diretoria

```
board_table = billing.join(quantity).join(ticket_average)
send_email(board_table, 'Todas as Lojas')
```
